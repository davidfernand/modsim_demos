## Monod Kinetic expression

We plot the results of using the Monod model. As ```functions``` folder does not have this function, we can implement it ourselves and plot the results.

````python
# We define the function that should return the specific growth rate directly
def Monod_kinetics(S,mu_max,Ks):
    '''
    S      : Substrate concentration
    mu_max : Maximum specific growth rate
    Ks     : Half-saturation constant
    '''
    return mu_max*(S/(Ks+S))
````

````python
# Parameters for the figure:
mu_max = 0.8 # h^-1
K_s = 5 # mg/L

S = np.linspace(0,50,num=50) #g/L

# We can plot the results of our model using the plotting library matplotlib (named plt)
fig = plt.figure(figsize=(6.5,5))
fig.subplots_adjust(hspace=0.2, wspace = 0.3)
ax1 = fig.add_subplot(111)
ax1.plot(S,Monod_kinetics(S,mu_max,K_s),'k-',label = r'$\mathrm{Monod}$')
ax1.plot([0,np.max(S)],[mu_max,mu_max],'k--',label = r'$\mathrm{\mu_{max}}$',alpha=0.75)
ax1.plot([0,K_s],[mu_max/2,mu_max/2],'b--',label = r'$\mathrm{\mu_{max}/2}$',alpha=0.75)
ax1.plot([K_s,K_s],[0,mu_max/2],'k-.',label = r'$\mathrm{K_s}$',alpha=0.75)
ax1.set_xlim([0,np.max(S)])
ax1.set_ylim([0,mu_max*1.11])
ax1.legend(loc=5,fontsize=15)
ax1.set_xlabel(r'$\mathrm{Substrate\/concentration,\/S[g/mL]}$', fontsize=20)
ax1.set_ylabel(r'$\mathrm{Specific\/growth\/rate,\/\mu_X[h^{-1}]}$', fontsize=20,)
fig.tight_layout(pad=0.4, w_pad=1, h_pad=1.0)
#plt.minorticks_on()

plt.show()
````

## Implementation of the Bioreactor model using the Monod growth model

We will now manually implement the ODE system and compare it to the one that we use when we import ```functions/monod```.

````python
# Python library to solve ODEs (and many more things!)
from scipy.integrate import odeint

# We define our model and we simultaneously solve it
def solve_Monod_model(t,u,mu_max,Y,Ks,Q,V,S_in,init):
    def model(u,t):
        '''
        Differential equation
        '''
        dXdt = Q/V*(-u[0])+mu_max*(u[1]/(Ks+u[1]))*u[0]          #Biomass X
        dSdt = Q/V*(S_in-u[1])-1/Y*mu_max*(u[1]/(Ks+u[1]))*u[0]  #Reactant S

        return [dXdt,dSdt]

    # We solve the system by providing the model, the initial conditions, and the time array
    y=odeint(model,init,t)
    
    return y
````

````python
# We provide the variables 
S_in = 0.001
S_init = 0.001
X_init = 1e-5

# Python list to be provided for the scipy solver
init = [X_init,S_init]

# Parameter values
Ks = 2e-4
mu_max = 3e-4
Q = 1
V  = 5000
Y = 0.5

# We define the time array with the start, final, and time step for the scipy integrator
t_final = 40
time = np.arange(0.,t_final*60*60,10)

# We solve the system
y = solve_Monod_model(time,init,mu_max,Y,Ks,Q,V,S_in,init)

# We store the solution
X = y[:,0]
S = y[:,1]

# We plot the solution
fig = plt.figure(figsize=(7.5,5))
fig.subplots_adjust(hspace=0.2, wspace = 0.3)

ax1 = fig.add_subplot(111)
ax1.plot(time/60/60,S,'b-',label = r'$\mathrm{[S]}$')
ax1.plot(time/60/60,X,'k-',label = r'$\mathrm{[X]}$')

ax1.plot(df.index/60/60,df['[S]'],'r--',label = r'$\mathrm{[S],functions/monod}$')
ax1.plot(df.index/60/60,df['[X]'],'g--',label = r'$\mathrm{[X],functions/monod}$')

ax1.legend(loc=1,fontsize=15)
ax1.set_ylabel(r'$\mathrm{C[g/mL]}$', fontsize=20)
ax1.set_xlabel(r'$\mathrm{t[h]}$', fontsize=20,rotation=0)
fig.tight_layout(pad=0.4, w_pad=1, h_pad=1.0)
ax1.grid(True)
#plt.minorticks_on()
plt.show()
````

## Linearisation of Monod kinetics around the operating point

We will now manually implement the linearised version of the bioreactor model using Monod kinetis (around the operating point) just like we did before and we will compare the results.


### 1st Operating point
If we directly substitute in the model in the first of the operating points  we get:
$$
\Big(\bar{X}_1 = 0 ; \bar{S}_1 = \bar{S}_{in}; \bar{S}_{in,1} = \bar{S}_{in} \Big)
$$

$$
\frac{d(X-\bar{X}_1)}{dt} = \frac{dX}{dt} \approx \Big(\mu_{max}\frac{\bar{S}_{in,1}}{K_{S}+\bar{S}_{in,1}} - \frac{Q}{V}\Big)(X-\bar{X}_1)
$$

$$
\frac{d(S-\bar{S}_1)}{dt} = \frac{dS}{dt} \approx  +\Big(-\frac{1}{Y}\mu_{max}\frac{\bar{S}_{in,1}}{K_{S}+\bar{S}_{in,1}}\Big)(X-\bar{X}_1) -\frac{Q}{V}(S-\bar{S}_{in,1}) + \frac{Q}{V}(\bar{S}_{in}-\bar{S}_{in,1})
$$

````python
# We implement different derivatives for the linearised version and solve it
def solve_Monod_lin_model_1(t,u,mu_max,Y,Ks,Q,V,S_in,init):
    def model(u,t):
        '''
        Differential equation
        '''
        X_op = 0
        S_op = S_in
        S_in_op = S_in

        # We build the ODE system with the deviation variables
        # Biomass X
        dXdt = (mu_max*(S_in/(Ks+S_in))-Q/V)*(u[0]-X_op) 
        # Substrate S
        dSdt = -(mu_max*S_in/(Y*(Ks+S_in)))*(u[0]-X_op) - Q/V*(u[1]-S_op) + Q/V*(S_in-S_in_op) 

        return [dXdt,dSdt]
        
    y=odeint(model,init,time)
    return y
````

### 2nd Operating point
If we directly substitute in the model in the second of the operating points we get:
$$
\Big(\bar{X}_2 = \frac{YQ(K_{S}+\bar{S}_{in})-Y\mu_{max}\bar{S}_{in}V}{Q-\mu_{max}V};\bar{S}_2 =\frac{K_{S}Q}{\mu_{max}V-Q};\bar{S}_{in,2} = \bar{S}_{in} \Big)
$$

$$
\frac{d(X-\bar{X}_2)}{dt} = \frac{dX}{dt} \approx +\Big(\frac{\mu_{max}\bar{S}_2}{K_{S}+\bar{S}_2} - \frac{Q}{V}\Big)(X-\bar{X}_2) + \frac{K_{S}\mu_{max}\bar{X}_2}{(K_{S}+\bar{S}_2)^2}(S-\bar{S}_2) 
$$

$$
\frac{d(S-\bar{S}_2)}{dt} = \frac{dS}{dt} \approx  +\Big(-\frac{\mu_{max}\bar{S}_2}{Y(K_{S}+\bar{S}_2)}\Big)(X-\bar{X}_2) + \Big(-\frac{K_{S}\mu_{max}\bar{X}_2}{Y(K_{S}+\bar{S}_2)^2}-\frac{Q}{V}\Big)(S-\bar{S}_{in,1}) + \frac{Q}{V}(\bar{S}_{in}-\bar{S}_{in,2})
$$

````python
# We implement different derivatives for the linearised version and solve it
def solve_Monod_lin_model_2(t,u,mu_max,Y,Ks,Q,V,S_in,init):
    def model(u,t):
        '''
        Differential equation
        '''
        # Definition of the second operating point
        X_op = (Y*Q*(Ks+S_in)-Y*mu_max*S_in*V)/(Q-mu_max*V)
        S_op = Ks*Q/(mu_max*V-Q)
        S_in_op = S_in

        # We build the ODE system with the deviation variables
        dXdt = (mu_max*S_op/(Ks+S_op)-Q/V)*(u[0]-X_op) + Ks*mu_max*X_op/(Ks+S_op)**2*(u[1]-S_op)                
        dSdt = (-mu_max*S_op/(Y*(Ks+S_op)))*(u[0]-X_op)  + (-Ks*mu_max*X_op/(Y*(Ks+S_op)**2) - Q/V)*(u[1]-S_op) + Q/V*(S_in-S_in_op)
                

        return [dXdt,dSdt]
        
    y=odeint(model,init,time)
    return y
````

````python
# Parameters, initial conditions, and time array
S_in = 0.001
S_init = 0.001
X_init = 1e-5
init = [X_init,S_init]
Q = 1
V  = 5000
Y = 0.5
Ks = 2e-4
mu_max = 3e-4
t_final = 45
time = np.arange(0.,t_final*60*60,10)

# We solve the correct model
y = solve_Monod_model(time,init,mu_max,Y,Ks,Q,V,S_in,init)
X = y[:,0]
S = y[:,1]

# We solve the linearise version
ylin = solve_Monod_lin_model_1(time,init,mu_max,Y,Ks,Q,V,S_in,init)
Xlin = ylin[:,0]
Slin = ylin[:,1]

ylin2 = solve_Monod_lin_model_2(time,init,mu_max,Y,Ks,Q,V,S_in,init)
Xlin2 = ylin2[:,0]
Slin2 = ylin2[:,1]

# We plot the results
fig = plt.figure(figsize=(7.5,5))
fig.subplots_adjust(hspace=0.2, wspace = 0.3)

ax1 = fig.add_subplot(111)
ax1.plot(time/60/60,S,'b-',label = r'$\mathrm{[S]}$')
ax1.plot(time/60/60,X,'k-',label = r'$\mathrm{[X]}$')

ax1.plot(time/60/60,Slin,'b--',label = r'$\mathrm{[S]_{lin,\/op. 1}}$')
ax1.plot(time/60/60,Xlin,'k--',label = r'$\mathrm{[X]_{lin,\/op. 1}}$')

ax1.plot(time/60/60,Slin2,'b-.',label = r'$\mathrm{[S]_{lin,\/op. 2}}$')
ax1.plot(time/60/60,Xlin2,'k-.',label = r'$\mathrm{[X]_{lin,\/op. 2}}$')

ax1.set_ylim(0,1.10*S_in)
ax1.legend(loc=1,fontsize=15)
ax1.set_ylabel(r'$\mathrm{C[g/mL]}$', fontsize=20)
ax1.set_xlabel(r'$\mathrm{t[h]}$', fontsize=20,rotation=0)
fig.tight_layout(pad=0.4, w_pad=1, h_pad=1.0)
#plt.minorticks_on()

plt.show()
````
